# Elon Musk - Analysis on Twitter account

Analyzing Elon Musk's Twitter data was an exhilarating journey! With columns like Datetime, Tweet ID, Text, and Username, this dataset provided a wealth of insights into his prolific tweets. Stay tuned as I unveil intriguing patterns and trends!

```python
import csv
```

```python
elonmusk = list(csv.reader(open("elonmusk.csv", encoding = "utf-8")))
```

# Remove TIME from datetime column and also create separate columns for date and time¬∂

```python
from datetime import datetime
```

```python
format_date = "%Y-%m-%d %H:%M:%S"

for i in elonmusk[1:]:
    i[0] = i[0].replace("TIME" , "")
    i[0] = datetime.strptime(i[0] , format_date)
```

```python
for i in elonmusk[0:]:
    print(i[0])
```

# Find at what time Elon Musk tweets the most !

```python
elonmusk[0][1]
```

```python
most_tweets = {}

for i in elonmusk[1:]:
    time = i[0]
    specific_time = time.strftime("%I %p")
    if specific_time not in most_tweets:
        most_tweets[specific_time] = 1
    else:
        most_tweets[specific_time] += 1
most_tweets 
```

```python
# Sort the dictionary based on values in descending order
sorted_most_tweets_desc = dict(sorted(most_tweets.items(), key=lambda item: item[1], reverse=True))

sorted_most_tweets_desc
```

# To Whom elon musk tweet most?

```python
freq_user = {}

for i in elonmusk[1:]:
    user = i[2]
    splited_user = user.split()
    
    if splited_user and splited_user[0].startswith("@"):
        freq_user[splited_user[0]] = freq_user.get(splited_user[0], 0) + 1
```

```python
freq_user
```

```python
most_freq_User = dict(sorted(freq_user.items(), key=lambda item: item[1], reverse=True))

most_freq_User
```
